```python
import pandas as pd

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5], 'B': [2, 4, 5, 4, 5]}
df = pd.DataFrame(data)

# Pearson Correlation
pearson_corr = df.corr(method='pearson')

# Spearman Correlation
spearman_corr = df.corr(method='spearman')

# Kendall Tau Correlation
kendall_corr = df.corr(method='kendall')

print("Pearson Correlation:\n", pearson_corr)
print("Spearman Correlation:\n", spearman_corr)
print("Kendall Tau Correlation:\n", kendall_corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap of the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

# Generating sample data
np.random.seed(0)
x = np.arange(1, 11)
y = x + np.random.normal(0, 1, 10)  # Adding some noise
```

```python

# Creating DataFrame
df_positive = pd.DataFrame({'x': x, 'y': y})
```

```python
df_positive
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df_positive['x'], df_positive['y'], color='b')
```

```python
# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df_positive['x'], df_positive['y'], color='b')

plt.title('Positive Correlation')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
np.random.seed(0)
x = np.arange(1, 11)
y = x + np.random.normal(0, 1, 10)  # Adding some noise

print(y)
```

```python
# Generating sample data
np.random.seed(0)
y_neg = -x + np.random.normal(0, 1, 10)  # Adding some noise

# Creating DataFrame
df_negative = pd.DataFrame({'x': x, 'y': y_neg})

# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df_negative['x'], df_negative['y'], color='r')
plt.title('Negative Correlation')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Generating sample data
np.random.seed(0)
y_none = np.random.normal(0, 1, 10)

# Creating DataFrame
df_no_corr = pd.DataFrame({'x': x, 'y': y_none})

# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df_no_corr['x'], df_no_corr['y'], color='g')
plt.title('No Correlation')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Generating sample data for multiple variables
np.random.seed(0)
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10) * 2
}

# Creating DataFrame
df = pd.DataFrame(data)

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# groupby

df1 = df[['body-style', 'drive-wheels', 'price']]
df1
```

```python
print(df1)
```

```python
df1.head()
```

```python
df1_group = df1.groupby(['body-style', 'drive-wheels'], as_index=True).mean()
df1_group
```

```python
# two categorial columns, body-style and drive-wheels
df1_group = df1.groupby(['body-style', 'drive-wheels'], as_index=False).mean()
df1_group
```

```python
df1_pivot = df1_group.pivot(index='drive-wheels', columns='body-style')
df1_pivot
```

```python
df1_pivot = df1_group.pivot(index='body-style', columns='drive-wheels')
df1_pivot
```

```python
#heatmap

import numpy as np


fig, ax = plt.subplots()
im = ax.pcolor(df1_pivot, cmap='RdBu')

#label names
row_labels = df1_pivot.columns.levels[1]
col_labels = df1_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df1_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df1_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
