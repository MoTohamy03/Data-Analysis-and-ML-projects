## **Data Normalization**

```python
# this is a new dataset we use it.

data = mydata[['width', 'length', 'height']]

data.head(15)
```

```python
mydata
```

```python
# this is a new dataset we use it.

data = mydata[['width', 'length', 'height']]

data.head(15)
```

```python
# simle feature scalling

data['width'] = data['width']/data['width'].max()
```

```python
# min-max >> This method scales the data to a fixed range, usually 0 to 1.
# It is done using the following formula: (value - min)/(max - min)

mydata['length'] = ( mydata['length']-mydata['length'].min() ) / ( mydata['length'].max()-mydata['length'].min())
mydata['length'].head(15)
```

```python
# z score >> This method scales the data based on the mean and standard deviation of the dataset.
# It centers the data around the mean with a standard deviation of 1.

mydata['height'] = ( mydata['height']-mydata['height'].mean() ) / mydata['height'].std()
mydata['height'].head(15)
```
